```python
import numpy as np
import scipy.signal.windows as windows

from matplotlib import pylab as pylab
from matplotlib import pyplot as plt

# %matplotlib inline
pylab.rcParams['figure.figsize'] = (8, 8)
```

```python
N = 16
```

```python
def verify(x):
    N2 = N // 2
    N4 = N // 4

    u = x[0:N4]
    v = x[N4:N2]

    vr = x[N2:3*N4]
    ur = x[3*N4:N]

    u2 = np.square(u)
    vr2 = np.square(vr)

    ur2 = np.square(ur)
    v2 = np.square(v)

    condition_1 = np.allclose(u2 + vr2, np.ones(N4))
    condition_2 = np.allclose(ur2 + v2, np.ones(N4))

    satisfied = condition_1 and condition_2



    print("#### x")
    print(x)

    print("#### u, v, vr, ur")
    print(u, v, vr, ur)

    print("#### u2, vr2, u2 + vr2")
    print(u2, vr2)
    print(u2 + vr2)

    print("#### ur2, v2, ur2 + v2")
    print(ur2, v2)
    print(ur2 + v2)

    print("#### Satisfies conditions")
    print(satisfied)

    print("#### Maxvalue")
    print(x.max())

    return satisfied
```

```python
wfltr = windows.barthann(16)

plt.figure()
plt.plot(wfltr)
plt.title(f"Bartlett-Hann window: 16. Satisified: {verify(wfltr)}")
```

```python
def vorbis(N=16):
    return np.sin(np.square(windows.cosine(N)) * np.pi/2)

wfltr = vorbis(16,)
plt.figure()
plt.plot(wfltr)
plt.title(f"Vorbis: 16. Satisified: {verify(wfltr)}")
```

```python
def iterbis(N, iter=1):
    if iter == 0:
        return windows.cosine(N)
    elif iter > 0:
        return np.sin(np.square(iterbis(N, iter-1)) * np.pi/2)

wfltr = iterbis(16, 3)
plt.figure()
plt.plot(wfltr)
plt.title(f"Iterbis: 16, iter=1. Satisified: {verify(wfltr)}")
```

```python
def iterbes(N, iter=1):
    if iter == 0:
        return windows.kaiser_bessel_derived(N, 6)
    elif iter > 0:
        return np.sin(np.square(iterbes(N, iter-1)) * np.pi/2)

wfltr = iterbes(16, 3)
plt.figure()
plt.plot(wfltr)
plt.title(f"Iterbes: 16, iter=3. Satisified: {verify(wfltr)}")
```
