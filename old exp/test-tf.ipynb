```python
import tensorflow as tf
import numpy as np
```

```python
# x = np.hstack([np.zeros(16), np.arange(64, dtype=np.float64), np.zeros(16)])

# x = np.arange(64, dtype=np.float64)

x = np.asarray([ 
    137, 154, 127, 107, 134, 148, 171, 228, 
    35,  199, 141, 238,  94, 179, 201,  90,
    213, 176, 226, 219, 212,  47, 250,  93,
    227, 226,  34, 220,  95,  36,  37, 162,
    184,  44,  94,  44, 182, 173,  39, 159,
     67,  98,  81, 122,  44, 104,  89, 184,
    178, 147, 225,  39, 176, 212, 113,  84,
    213, 243, 193,  56, 183,   2,  69, 212,
], dtype=np.float64)

x = x - 128
```

```python
x
```

```python
y = tf.signal.mdct(x, frame_length=16, window_fn=tf.signal.kaiser_bessel_derived_window, norm='ortho')
```

```python
npy = np.asmatrix(y).copy()

npy[:, -5:] = 0

print(npy)
```

```python
z = tf.signal.inverse_mdct(npy, window_fn=tf.signal.kaiser_bessel_derived_window, norm='ortho')

print(z)
```

```python
px = np.reshape(x, (8, 8))
pz = np.asarray(z).reshape((8, 8)).copy()
for i in range(8):
    print(f'-----{i}-----')
    print(px[i, :] + 125)
    print(pz[i, :] + 125)
```
